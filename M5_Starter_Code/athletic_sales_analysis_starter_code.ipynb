```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv', low_memory=False)
sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv', low_memory=False)
```

```python
# Display the 2020 sales DataFrame
sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
merged_data = pd.concat([sales_2020,sales_2021], axis=0)
merged_data.reset_index(drop=True, inplace=True)
merged_data.head()
```

```python
# Check if any values are null.
merged_data.isnull().sum()
```

```python
# Check the data type of each column
merged_data.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
merged_data['invoice_date'] = pd.to_datetime(merged_data['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
merged_data.head()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold = merged_data[['region', 'state', 'city', 'units_sold']]
products_sold = products_sold.groupby(['region','state','city'])[['units_sold']].agg(['sum'])
products_sold = products_sold.rename(columns={'sum': 'Total_products_Sold'})
# Show the top 5 results.
products_sold.head()                  
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_pivot = pd.pivot_table(products_sold,
                               values='units_sold',
                               index=['region', 'state', 'city']
                              )

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_pivot = products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
products_sold_pivot.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = merged_data[['region', 'state', 'city', 'total_sales']]
total_sales = total_sales.groupby(['region','state','city'])[['total_sales']].sum()
total_sales = total_sales.rename(columns={'total_sales': 'Total Sales'})
# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = pd.pivot_table(total_sales,
                                   values='Total Sales',
                                   index=['region', 'state', 'city'],
                                  )

# Optional: Rename the "total_sales" column to "Total Sales"

# Show the top 5 results.
total_sales_pivot.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_retailer = merged_data[['retailer', 'region', 'state', 'city', 'total_sales']]
total_sales_retailer = total_sales_retailer.groupby(['retailer', 'region','state','city'])[['total_sales']].sum()
total_sales_retailer = total_sales_retailer.rename(columns={'total_sales': 'Total Sales'})
# Show the top 5 results.
total_sales_retailer.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailer_pivot = pd.pivot_table(total_sales_retailer,
                                            values='Total Sales',
                                            index=['retailer', 'region', 'state', 'city']
                                           )

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
total_sales_retailer_pivot.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_footwear_data = merged_data.loc[merged_data['product'] == "Women's Athletic Footwear"]
women_footwear_data.reset_index(drop=True, inplace=True)
women_footwear_data.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
number_sold_women = women_footwear_data[['retailer', 'region', 'state', 'city', 'units_sold']]
number_sold_women = number_sold_women.groupby(['retailer', 'region','state','city'])[['units_sold']].sum()
number_sold_women = number_sold_women.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
# Show the top 5 results.
number_sold_women.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
number_sold_women_pivot = pd.pivot_table(number_sold_women,
                                         values='Womens_Footwear_Units_Sold',
                                         index=['retailer', 'region', 'state', 'city']
                                        )

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.
number_sold_women_pivot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
date = pd.pivot_table(women_footwear_data,
                      values='total_sales',
                      index='invoice_date'
                     )

# Optional: Rename the "total_sales" column to "Total Sales"
date = date.rename(columns={'total_sales': 'Total Sales'})

# Show the table.
date.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_total_sales = date.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_total_sales.sort_values(by="Total Sales", ascending=True).head()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_total_sales = date.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_total_sales.sort_values(by="Total Sales", ascending=True).head()
```
